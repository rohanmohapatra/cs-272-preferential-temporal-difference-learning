```python
from linearfa_agent import *
import numpy as np
import random

random.seed(200)

params = np.arange(0, 1, 0.2).tolist()
```

```python
sarsa_returns = []
sarsalambda_returns = []
ptd_returns = []

sarsa_rmse = []
sarsalambdasarsa_rmse = []
ptd_rmse = []
for each in params:
    environment = gym.make('CartPole-v1')
    sarsa_agent = SARSAAgent(environment, 4)
    sarsa_agent.learn()
    episode, done = sarsa_agent.best_run()
    sarsa_returns.append(sarsa_agent.calc_return(episode, done))
    sarsa_rmse.append(sarsa_agent.rmse())

    environment = gym.make('CartPole-v1')
    sarsa_lambda_agent =  SARSALambdaAgent(environment, 4, each)
    sarsa_lambda_agent.learn()
    episode, done = sarsa_lambda_agent.best_run()
    sarsalambda_returns.append(sarsa_lambda_agent.calc_return(episode, done))
    sarsalambdasarsa_rmse.append(sarsa_lambda_agent.rmse())

    environment = gym.make('CartPole-v1')
    ptd_agent = PTDAgent(environment, 4, each)
    ptd_agent.learn()
    episode, done = ptd_agent.best_run()
    ptd_returns.append(ptd_agent.calc_return(episode, done))
    ptd_rmse.append(ptd_agent.rmse())
    print("------------------------------")
```

```python
import matplotlib.pyplot as plt

plt.plot(params, sarsa_returns, label="SARSA")
plt.plot(params, sarsalambda_returns, label="SARSA Lambda")
plt.plot(params, ptd_returns, label="PTD")
plt.xlabel("Params (could be lamba or beta based on algorithm)")
plt.ylabel("Returns")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
print(ptd_rmse)

plt.plot(params, sarsa_rmse, label="SARSA")
plt.plot(params, sarsalambdasarsa_rmse, label="SARSA Lambba")
plt.plot(params, ptd_rmse, label="PTD")
plt.xlabel("Params (could be lamba or beta based on algorithm)")
plt.ylabel("Returns")
plt.xticks(params)
plt.legend()
plt.show()
```
